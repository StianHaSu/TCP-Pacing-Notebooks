```python
RENDER_SINGLE_GRAPHS = True
RENDER_COMBINED_GRAPHS = True
```

```python
def get_data_points(filename: str):
    data = []
    with open(filename, "r+") as f:
        for line in f:
            data.append(int(line.strip()))
            
    return data
```

## Linux reno 125 BDP

```python
from experiments.max_flightsize.plotting import plot_ssthresh_and_queue_size

## Reno in flight 

reno_no_pacing_125bdp_in_flight = get_data_points("inflight_reno.txt")
reno_default_pacing_125bdp_in_flight = get_data_points("normal_reno_pacing.txt")
reno_log_pacing_125bdp_in_flight = get_data_points("exp_20260213-161904_inflight.txt")

queues = [ x for x in range(5, 255, 5) ]
```

```python
if RENDER_SINGLE_GRAPHS:
    plot_ssthresh_and_queue_size(queues, 125, reno_no_pacing_125bdp_in_flight, None, "Max packets in flight NewReno 125BDP", color="red", save=True)
    plot_ssthresh_and_queue_size(queues, 125, reno_default_pacing_125bdp_in_flight, None, "Max packets in flight NewReno default pacing 125BDP", color="green", save=True)
    plot_ssthresh_and_queue_size(queues, 125, reno_log_pacing_125bdp_in_flight, None, "Max packets in flight NewReno log pacing 125BDP", color="blue", save=True)
```

```python
from experiments.max_flightsize.plotting import plot_ssthresh_and_queue_size_multiple

if RENDER_COMBINED_GRAPHS:
    plot_ssthresh_and_queue_size_multiple(queues,
                                          reno_no_pacing_125bdp_in_flight,
                                          reno_default_pacing_125bdp_in_flight,
                                          reno_log_pacing_125bdp_in_flight,
                                          "NewReno",
                                          125,
                                          "Max_flight_size_reno_125bdp",
                                      save=True)
```
