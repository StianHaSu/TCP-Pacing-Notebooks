## Logarithmic pacing

```python
bandwidths = [50, 100, 250, 500, 750, 1000]
queues = [10, 25, 50, 75, 100, 250, 500]

result = transform_to_matrix("data/exp_20260305-192159_inflight-logarithmic.txt", bandwidths, queues)

percentage = get_as_percentage_matrix(result, bandwidths, queues, 30).round().astype(int)
percentage = percentage.clip(upper=100)
plot_heatmap(percentage, "Reno with logarithmic pacing")
```

## No Pacing

```python
import numpy as np

bandwidths = [50, 100, 250, 500, 750, 1000]
queues = [10, 25, 50, 75, 100, 250, 500]

aggregate = aggregate_data("data/exp_20260305-231449_inflight-none.txt", 10)
statistics = get_statistics(aggregate)

mean_values = transform_to_matrix_mean(statistics.iloc[:, 0].tolist(), bandwidths, queues)
stddev_values = transform_to_matrix_stddev(statistics.iloc[:, 1].tolist(), bandwidths, queues)

stddev_percentage = stddev_values.div(mean_values).mul(100)
stddev_percentage = stddev_percentage.replace([np.inf, -np.inf], 0).fillna(0)

percentage = get_as_percentage_matrix(mean_values, bandwidths, queues, 30).round().astype(int)
percentage = percentage.clip(upper=100)
plot_heatmap(percentage, "appendix heatmap mean max flight one queue no pacing")
plot_heatmap_stddev(stddev_percentage, "appendix heatmap stddev max flight one queue no pacing")
```
